## 1. Print numbers from 0 to 9

```python
for i in range(10):
    print(i)
```

## 2. Print numbers from 1 to 20

```python
for i in range(1, 21):
    print(i)
```

## 3. Print even numbers from 1 to 20

```python
for i in range(2, 21, 2):
    print(i)
```

## 4. Print odd numbers from 1 to 20

```python
for i in range(1, 21, 2):
    print(i)
```

## 5. Print numbers from 10 to 1 (reverse)

```python
for i in range(10, 0, -1):
    print(i)
```

## 6. Sum of first N natural numbers

```python
n = 5
total = 0
for i in range(1, n + 1):
    total += i
print("Sum:", total)
```

## 7. Multiplication table of 7

```python
num = 7
for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

## 8. Count numbers divisible by 3 between 1 and 50

```python
count = 0
for i in range(1, 51):
    if i % 3 == 0:
        count += 1
print("Count:", count)
```

## 9. Print squares from 1 to 10

```python
for i in range(1, 11):
    print(i * i)
```

## 10. Print list elements using index

```python
fruits = ["apple", "banana", "mango"]
for i in range(len(fruits)):
    print(i, fruits[i])
```
